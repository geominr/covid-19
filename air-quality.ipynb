```python
import pandas as pd; import requests; from io import StringIO; import time
ts = pd.read_csv('covid-county-data/covid-deaths.csv')
static = pd.read_csv('covid-county-data/static.csv')
ts = pd.merge(static[['FIPS','Lat','Long_']], ts, on='FIPS', how='left')
days = [x.split('-')[2]+'-'+x.split('-')[0]+'-'+x.split('-')[1] for x in ts.columns[3:]]
cols = ts.columns[3:]
reqs = 0
#print("Waiting one hour starting at 11:20PM.....")
#time.sleep(60*59)
for i in ts.index:
    print("")
    print(static.Combined_Key[i]+": ", end="")
    url0 = 'http://www.airnowapi.org/aq/observation/latLong/historical/?format=text/csv&'
    url1 = 'latitude={}&longitude={}'.format(ts.Lat[i], ts.Long_[i])
    url3 = '&distance=25&API_KEY=7CED8FC4-2DCE-47F4-BA3A-0F3EC1DAF1B0'
    for d in range(0,len(days)):
        day = days[d]
        url2 = '&date={}T00-0000'.format(day)
        url = url0+url1+url2+url3
        if reqs < 499:
            r = requests.get(url)
            reqs += 1
            df = pd.read_csv(StringIO(r.text), sep=',')
            if len(df) > 0:
                ts.at[i, cols[d]] = df["AQI"][0]
                print("0", end="")
            else:
                ts.at[i, cols[d]] = "NAN"
                print("X",end="")
        else:
            print("Waiting......", end="\r")
            time.sleep(60*60)
            r = requests.get(url)
            reqs += 1
            df = pd.read_csv(StringIO(r.text), sep=',')
            if len(df) > 0:
                ts.at[i, cols[d]] = df["AQI"][0]
                print("0", end="")
            else:
                ts.at[i, cols[d]] = "NAN"
                print("X",end="")
            reqs = 0
            
ts.to_csv('covid-county-data/daily-aqi.csv')
```
